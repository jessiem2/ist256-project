```python
import requests

apikey = '58846e1b2ea8652926535729c9d3b4f4'
country = 'spain'
url = f'https://ws.audioscrobbler.com/2.0/?method=geo.gettoptracks&country={country}&api_key={apikey}&format=json&limit=5'

response = requests.get(url)
content = response.json()
content
```

```python
import folium

center = (0, 0)
country_geojson = 'https://raw.githubusercontent.com/datasets/geo-countries/master/data/countries.geojson'
map = folium.Map(location=center, zoom_start=2)
folium.Choropleth(geo_data=country_geojson, data=content, columns=['artist', 'name'],
                  key_on ='feature.id', fill_color='YlOrRd', legend_name='Songs').add_to(map)

map

for row in content.to_records():
    pos = (row['track'],row['duration'])
    marker = folium.Marker(location=pos, 
                    popup="Song: %s" % (row['duration'])
                          )
    map.add_child(marker)
map
```
